```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ORIGINAL_BED_FILE = "~/Desktop/D23_Col0_all_CpG.bed"
SAMPLE_BED_FILE = "~/Desktop/D23_Col0_all_CpG_sample_0_ray.bed"

col_labels = ["chr", "start", "end", "uc", "mc", "prop"]
original_bed_data = pd.read_csv(ORIGINAL_BED_FILE, sep='\t', names=col_labels, header=None)
sample_bed_data = pd.read_csv(SAMPLE_BED_FILE, sep='\t', names=col_labels, header=None)
```

```python
def plot_hist(bed_data):
    read_counts = bed_data["uc"] + bed_data["mc"]
    x_min = read_counts.min()
    x_max = read_counts.max()
    
    n_bins = 20
    
    boundary_0 = x_min
    boundary_1 = 40
    boundary_2 = 50 
    boundary_3 = 275
    boundary_4 = 2000
    boundary_5 = 10000
    boundary_6 = x_max
    
    fig, axs = plt.subplots(nrows=2, ncols=3, sharey=False, sharex=False, tight_layout=True, figsize=(15, 10))
    
    axs[0, 0].hist(read_counts, range=(boundary_0, boundary_1), bins=n_bins)
    axs[0, 0].set_title(f"{boundary_0} - {boundary_1} reads")
    axs[0, 1].hist(read_counts, range=(boundary_1, boundary_2), bins=n_bins)
    axs[0, 1].set_title(f"{boundary_1} - {boundary_2} reads")
    axs[0, 2].hist(read_counts, range=(boundary_2, boundary_3), bins=n_bins)
    axs[0, 2].set_title(f"{boundary_2} - {boundary_3} reads")
    axs[1, 0].hist(read_counts, range=(boundary_3, boundary_4), bins=n_bins)
    axs[1, 0].set_title(f"{boundary_3} - {boundary_4} reads")
    axs[1, 1].hist(read_counts, range=(boundary_4, boundary_5), bins=n_bins)
    axs[1, 1].set_title(f"{boundary_4} - {boundary_5} reads")
    axs[1, 2].hist(read_counts, range=(boundary_5, boundary_6), bins=n_bins)
    axs[1, 2].set_title(f"{boundary_5} - {boundary_6} reads")
    
    plt.show()
```

```python
plot_hist(original_bed_data)
```

```python
plot_hist(sample_bed_data)
```
